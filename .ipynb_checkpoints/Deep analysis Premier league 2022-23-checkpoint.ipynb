# Exploratory data analysis

## *Which is the strongest team at home?* 

First, let's create a new column called 'Win At Home'. This new column will help in calcuate several things more easily later. This column will be a category type. This column will have 3 options. Win, Lose or Tie. 

```python
# First, let's create a function with the possible results
def did_local_team_win(goals_home, away_goals):
    if goals_home>away_goals:
        return 'Win'
    elif goals_home<away_goals:
        return "Lose"
    elif goals_home==away_goals:
        return "Tie"
        
premier['Win At Home'] = premier.apply(lambda row: did_local_team_win(row['Goals Home'], row['Away Goals']), axis=1)
# Now, let's become this new column in a category column. 
premier['Win At Home'] = premier['Win At Home'].astype('category').cat.codes
```

```python
# Now, let's check how the column looks. 
premier[['Home Team', 'Goals Home', 'Away Team', 'Away Goals', 'Win At Home']].head(7)
```

Ok, so now we know that the categories in the 'Win At Home' colummn look like this:
- When local team wins = 2
- When local team tie = 1
- When local team lose = 0

It's moment to create the first visualizations. Let's check which teams have the highest number of victories, draws and losses playing as local. 

```python
more_victories_home = premier[premier['Win At Home']==2].groupby(premier['Home Team'])['Win At Home'].count().reset_index().rename(
    columns = {"Win At Home":"Number of victories at home"}).sort_values("Number of victories at home", ascending=False)
more_victories_home = more_victories_home.reset_index(drop=True).style.background_gradient()
more_victories_home
```

```python
more_draws_home = premier[premier['Win At Home']==1].groupby(premier['Home Team'])['Win At Home'].count().reset_index().rename(
    columns = {"Win At Home":"Number of draws at home"}).sort_values("Number of draws at home", ascending=False)
more_draws_home = more_draws_home.reset_index(drop=True)
more_draws_home.head(7)
```

```python
more_losses_home = premier[premier['Win At Home']==0].groupby(premier['Home Team'])['Win At Home'].count().reset_index().rename(
    columns = {"Win At Home":"Number of losses at home"}).sort_values("Number of losses at home", ascending=False)
more_losses_home = more_losses_home.reset_index(drop=True)
more_losses_home.head(7)
```

<b>Conclusions:</b>
- As we saw, the three teams with more victories were Manchester City, Manchester United, and Arsenal. None of these teams have a high number of losses or draws; that makes sense because a high number of victories at home is a sign of a good performance.
- These teams had the most points gathered at the end. We can conclude that having good results at home is related to finishing upper in the competition.
- In fact, two of the teams that had the most defeats at home ended up being relegated. The other team (Everton) ended up in 17th place, very close to being relegated. 
- Something interesting to note is that even though Manchester United is the second team with more victories at home, they finished in third place under Arsenal in the competition.

## How do England look by points?

As you probably know, soccer leagues are based on points. That means, the team who has earned the most points would be the winner. In the 2022-23 season that team was the Manchester city with 89 points, and the runners-up were Arsenal and Manchester United with 84 and 75 respectively. Anyway, we can tell that just seeing the standing table of that season, but a really interesting question would be: which regions of England are the strongest in the competition? and why? In order to solve that questions, let's analysis a graph regarding the number of points that those regions accumulate. 

```python
# First, we need to make a dictionary that can tell us from which region is every team.
england_team_regions_dict = {
    'Arsenal':'London', 'Aston Villa':'West Midlands', 'Bournemouth':'South West', 
    'Brentford':'London', 'Brighton and Hove Albion':'South East', 'Chelsea':'London', 
    'Crystal Palace':'London', 'Everton':'North West', 'Fulham':'London', 
    'Leeds United':'Yorkshire and the Humber', 'Leicester City':'East Midlands', 'Liverpool':'North West', 
    'Manchester City':'North West', 'Manchester United':'North West', 'Newcastle United':'North East', 
    'Nottingham Forest':'East Midlands', 'Southampton':'South East', 'Tottenham Hotspur':'London', 
    'West Ham United':'London', 'Wolverhampton Wanderers':'West Midlands'
}
# Now, let's create a new column called 'Region' with this information.
# I will make it regarding who played as local that day. 
premier['Region'] = premier['Home Team'].replace(england_team_regions_dict)
premier[['Home Team', 'Region']].head(7)
```

The next step is create a new dictionary of regions and how many points those regions have. To do that, I will check each match and check if the home team won. If the home team won the region it would add 3 points, if it tied 1, and if it lost 0. 

```python
regions_dict = {
    'London':0, 'North East':0, 'North West':0, 
    'South East':0, 'South West':0, 'East Midlands':0, 
    'West Midlands':0, 'East of England':0, 'Yorkshire and the Humber':0
}
# Now, let's create a function to add this points.
# I will use the column (Win_At_Home) created in the section 'Which is the strongest team at home?' to check who won
def region_calculator(dict, region_dict, win_at_home, home_team):
    # Only is necessary make conditions for win and draw because lose add 0 points. 
    if win_at_home==2:
        region = region_dict[home_team]
        dict[region] += 3
    elif win_at_home==1:
        region = region_dict[home_team]
        dict[region] += 1

# Now, let's use the function.
premier.apply(lambda row: region_calculator(regions_dict, england_team_regions_dict, row['Win At Home'], row['Home Team']), axis=1)
```

## Big 6 performance

If you are a fan of the english football, probably you know that the Premier league is dominated for 6 big teams. These teams are Arsenal, Manchester City, Manchester United, Liverpool, Spurs, and Chelsea. The six teams are known as the Big 6. For this section we are going to take a closer look of the performances of these teams, and compare their performances from date 1 to the final date 38.  

To do that, let's create a new dictionary with two list. The first will be the dates (1-38) and the second will be how many points the teams had accumulated up to that date. 

```python
# Dictionary creation.
bix_6_performance = {
    "dates": ['date 0'],
    "Arsenal": [0],
    "Chelsea": [0],
    "Liverpool": [0],
    "Manchester City": [0],
    "Manchester United": [0],
    "Tottenham Hotspur": [0],
}
# now we complete the list with the remaining dates.
def complete_dict(dict):
    for date_number in range(1, 39):
        dict["dates"].append('date %s' % date_number)
    return dict
bix_6_performance = complete_dict(bix_6_performance)
# A list with the names of the big 6 is necessary too. 
big_6 = ['Arsenal', 'Chelsea', 'Liverpool', 'Manchester City', 'Manchester United', 'Tottenham Hotspur']
```

Next step is use the list 'big_6' and the 'Win At Home' column created in the section 'Which is the strongest team at home?' to filter the results of these teams in the data set. Every time they have won beign locals or the local team has lost when their were visitors they will add 3 points, 1 in case of draw and 0 for loss. The sum of points will change on each date, for example, if they won on the date 1 they would have 3 points for date 1, then if they tied on date 2 they would have 4 points for that date, and then if they lost on date 3 they would have 4 points for that date. 

```python
# Let's filter for the big 6
premier_b6 = premier[['Home Team', 'Goals Home', 'Away Team', 'Away Goals', 'Win At Home']]
premier_b6 = premier[premier['Home Team'].isin(big_6) | premier['Away Team'].isin(big_6)]
# Let's check the lenght due confirm the filter worked good. The lenght should be 198.
# If each of them play 38 games per season that is (38 * 6), on that substract the times they played each other that is (5 * 6)
# (19 * 6) - (5 * 6) = 198
len(premier_b6)
```

```python
# Now is moment to fill the dictionay's 'points' list.
def count_points(home_team, away_team, win_at_home, big_6, dict):
    # In the scenarios when two teams of the big 6 face each other. 
    if (home_team in big_6) and (away_team in big_6):
        if win_at_home == 2:
            dict[home_team].append(dict[home_team][-1] + 3)
            dict[away_team].append(dict[away_team][-1] + 0)
        elif win_at_home == 1:
            dict[home_team].append(dict[home_team][-1] + 1)
            dict[away_team].append(dict[away_team][-1] + 1)
        elif win_at_home == 0:
            dict[home_team].append(dict[home_team][-1] + 0)
            dict[away_team].append(dict[away_team][-1] + 3)
    elif home_team in big_6:
        # In the scenarios when one team in the big 6 is playing as local.
            if win_at_home == 2:
                dict[home_team].append(dict[home_team][-1] + 3)
            elif win_at_home == 1:
                dict[home_team].append(dict[home_team][-1] + 1)
            elif win_at_home == 0:
                dict[home_team].append(dict[home_team][-1] + 0)
        # In the scenarios when one team in the big 6 is playing as visitor.
    elif away_team in big_6:
            if win_at_home == 0:
                dict[away_team].append(dict[away_team][-1] + 3)
            elif win_at_home == 1:
                dict[away_team].append(dict[away_team][-1] + 1)
            elif win_at_home == 2:
                dict[away_team].append(dict[away_team][-1] + 0)
    return dict
```

```python
premier_b6.apply(lambda row: count_points(row['Home Team'], row['Away Team'], row['Win At Home'], big_6, bix_6_performance), axis=1)
```

## Expected goals per match

One of the most common uses of data science in the sports world is to make bets. In soccer, there is a popular type of bet about the expected goals per match. You can bet that in a match there will be more or less <b>n</b> number of goals. For example, for the match Chelsea vs. Newcastle United on May 28, 2023, you could bet that the expected goals for that match will be less than 2.5 goals, so if at the end of the match between the two teams they scored less or equal to 2 goals, you will win the bet, but if they scored more than 2 goals, you will lose. In this example, the match finished 1-1, so to win it, you should bet less than 2.5 goals.

In this section, I will create an expected goals calculator. In order to create the calculator, I will use the Poisson distribution. Recall that the Poisson distribution is often used to model the number of events that occur within a fixed interval of time or space. For example, if you want to know what the probability is that in the next Arsenal game they will score more than 3.5 goals, you can use the Poisson distribution to check the probabilities that happen.

```python
# First, I need to import scipy.stats poisson. 
from scipy.stats import poisson
# and other libraries to create visual input.
import ipywidgets as widgets
from IPython.display import display
```

```python
# The next step is to create a function that can calculate the number of goals scored and received by the chosen team this season. 
def avg_goals(team_name, home_team, away_team, goals_home, away_goals, lambda_parameter):
    if (team_name == home_team) | (team_name == away_team):
        lambda_parameter['lambda_parameter'] = lambda_parameter['lambda_parameter'] + goals_home + away_goals
```

```python
# Now let's make an interface were the user can pick a team. 
# First, it's necessary to have a list with the names of all the teams.
# I already created one in the section 'Cleaning and Validation'. 
# In that section, the variable became in a set, so it's necessary to make it a list again.
teams_2022_23 = list(teams_2022_23)
# and I am going to sort it alphabetically.
teams_2022_23 = sorted(teams_2022_23)
```

```python
# Create the dropdown widget
dropdown = widgets.Dropdown(options=teams_2022_23, value=teams_2022_23[0], description='Select a team:')
# Display the dropdown widget
display(dropdown)
```

```python
# Create the input to ask the number of goals.
expected_goals = input('How many goals do you think will be scored in the next %s match? ' %dropdown.value)
```

```python
# Next, we need to calculate the average of goals per match for the picked team.
lambda_parameter = {'lambda_parameter': 0}
# and we are going to use the function created before. 
premier.apply(lambda row: avg_goals(str(dropdown.value), row['Home Team'], row['Away Team'], row['Goals Home'], 
                                    row['Away Goals'], lambda_parameter), axis=1)
# lambda parameter = (scored_goals + received_goals)/(number_of_matches_in_the_season)
lambda_parameter = lambda_parameter['lambda_parameter']/38
```

```python
# Finally, we apply the Poisson distribution.
expected_goals = int(expected_goals)
pmf_values = poisson.pmf(expected_goals, lambda_parameter)
print("There is a " + str(round(pmf_values*100, 2)) + "% probability that " + dropdown.value + 
      " finish their next game with this number of goals.")
print("Recall that this is the number of goals scored for both teams, not just for %s." % dropdown.value)
```

<b>Note: </b>There is a way to make this percentage more accurate. That would be using all the recent matches of the match you are going to bet on. For example, you could use the Poisson distribution in the same way in any Chelsea vs. Newcastle United match, but if you only used the data of all the past (at least the more recent) matches between these two teams, discarding any other match, the percentage would be more accurate.

<b>Note: </b>In this notebook, it wasn't made because it only had data from the season 2022–23, so for any confrontation, the distribution would only have the data of two matches to predict, so it's a small sample. Anyway, here is the implementation of that. In the future, I'm planning to add the data from other seasons and make this percentage more accurate.

```python
# Let's create a function that can calculate the number of goals by the chosen combination this season. 
def avg_goals_2(team_name_home, team_name_away, home_team, away_team, goals_home, away_goals, lambda_parameter):
    if ((team_name_home == home_team) & (team_name_away == away_team)) | ((team_name_home == away_team) & (team_name_away == home_team)):
        lambda_parameter['lambda_parameter'] = lambda_parameter['lambda_parameter'] + goals_home + away_goals
```

```python
# Create the dropdown widget
dropdown_home = widgets.Dropdown(options=teams_2022_23, value=teams_2022_23[0], description='Home team:')
dropdown_away = widgets.Dropdown(options=teams_2022_23, value=teams_2022_23[1], description='Away team:')
# Display the dropdown widget
display(dropdown_home)
display(dropdown_away)
```

```python
# Create the input to ask the number of goals.
expected_goals = input('How many goals do you think will be scored in the '+ dropdown_home.value +
                       ' vs. '+ dropdown_away.value +' match? ')
```

```python
# Next, we need to calculate the average of goals in this match for the picked teams.
lambda_parameter = {'lambda_parameter': 0}
# and we are going to use the function created before. 
premier.apply(lambda row: avg_goals_2(str(dropdown_home.value), str(dropdown_away.value), row['Home Team'], row['Away Team'], row['Goals Home'], 
                                    row['Away Goals'], lambda_parameter), axis=1)
# lambda parameter = (scored_goals + received_goals)/(number_of_matches_in_the_season)
lambda_parameter = lambda_parameter['lambda_parameter']/2
```

```python
expected_goals = int(expected_goals)
pmf_values = poisson.pmf(expected_goals, lambda_parameter)
```

```python
# Finally, we apply the Poisson distribution.
expected_goals = int(expected_goals)
pmf_values = poisson.pmf(expected_goals, lambda_parameter)
print("There is a " + str(round(pmf_values*100, 2)) + "% probability that this match finish with " + str(expected_goals) + " goals.")
```
